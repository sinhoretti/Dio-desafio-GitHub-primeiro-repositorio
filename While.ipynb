```python
#Escreva um sistema que lei a senha e compare se é verdadeira!
#Se estiver errada pede digitar a senha novamente!
contador=0
senha = "senai"
while contador==0:
  senha = input(print("\nDigite sua senha: "))
  if(senha == "senai"):
    print("Senha Correta! Acesso Liberado!")
    contador=1
    break
  else:
    print("Senha incorreta! Acesso Negado!")
    contador=0
```
